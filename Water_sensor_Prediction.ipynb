```python
def get_cols_zero_std(df: pd.DataFrame):
    cols_to_drop = []
    num_cols = [ i for i in df.columns[1:] if df[i].dtype != "0"]
    for i in df.columns[1:]:
        if df[i].std() == 0:
            cols_to_drop.append(i)

    return cols_to_drop
```

```python
def get_reduntant_col(df: pd.DataFrame, missing_thresh = 0.7):
    cols_missing_ratio = df.isnull().sum().div(df.shape[0])
    cols_to_drop = list(cols_missing_ratio[cols_missing_ratio > missing_thresh].index)

    return cols_to_drop

    

```

```python
cols_drop_1 = get_cols_zero_std(df = df)
cols_drop_1
```

```python
cols_drop_2 = get_reduntant_col(df, missing_thresh= .7)
cols_drop_2
```

```python
cols_to_drop = cols_drop_1 + cols_drop_2 + ['Wafers']
len(cols_to_drop)
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline
import pandas as pd

# Assuming df is your DataFrame
data = {'Feature1': [1, 2, 3, None, 5],
        'Feature2': [5, None, 15, 20, 25]}
df = pd.DataFrame(data)

# Set up the imputer
imputer = KNNImputer(n_neighbors=3)

# Set up the scaler
scaler = RobustScaler()

# Create the pipeline
prep_process = Pipeline(steps=[
    ('imputer', imputer),
    ('scaling', scaler)
])

# Fit and transform the data using the pipeline
transformed_data = prep_process.fit_transform(df)

# Display the transformed data
print("Original Data:")
print(df)
print("\nTransformed Data:")
print(transformed_data)
```
